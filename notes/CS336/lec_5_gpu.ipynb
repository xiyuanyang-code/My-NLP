## GPU in Depth

### CPUs and GPUs

- CPU 需要强大的计算能力来执行程序生成的二进制代码和汇编指令，需要强大的控制器，并且需要较好的分支预测能力
- GPU 相比于 CPU 具有很多并行的 ALU 算数运行单元，控制器在 CPU 负责协调调度 GPU 的多个 ALU 计算单元。
    - 程序的并行和**吞吐量**是 GPU 运行控制的重点之一

### Atanomy of GPUs

- GPU 存在大量的 SM (Streaming MultiProcessors), 每一个都可以独立的调度多个 SPs (Streaming Processor)，这些可以实现超高进程的并发执行。

    - SP (CUDA Core): 代表并行执行的最基本单元，每一个 SP 可以在一个时钟周期内执行一个时间的计算
    - SM: SM 是由多个 SP 加上其他资源组成的高级处理单元

GPU 同样满足局部性原理，在 SM 内部的 L1 & shared memory 运行速度非常快. (The closer the memory to the SM, the faster it is.)



## Execution Models of GPUs

- Blocks: groups of threads. 每一个 block 运行在一个 SM 上，并有自己独享的内存 (就像一个线程!)
- Threads: all threads execute the same instructions but with different inputs (SIMT).
- Warp: Threads always execute in a 'warp' of 32 consecutively numbered threads each. 一个 Warp 固定包含连续编号的 32 个 thread，GPU 的指令获取单元一次只取一条指令，然后这条指令会同时分发给 Warp 内的所有 32 个线程。

SM 的调度以 Warp 作为基本单元进行构建，而不会单独的调度某个单个线程。

## Memory Models of a GPU

和 CPU 一样，GPU 也存在分层的内存结构：

- Shared Memories in a single block
    - 在一个 Block 中的不同 thread 可以快速的读取当前的 shared memory
- Global Memory 和 Constant Memory 可以存储更多的数据，但是运行访问数据的速度会变慢


## TPUs

GPUs, TPUs, and many other accelerators are at a high level, similar.

- Scalar Unit: dispatching instructions to the VPU and MXU
- Vector Unit: VPU
- Matrix Multiply Unit: MXU
- High Bandwith Memory: HBM

Tensor Core 被极度优化方便做快速的矩阵乘法。


- Easily scale up hard workloads (by adding more SMs).
- SIMT 机制
- Threads 非常轻量化, can be stopeed and started easily.

GPU 在硬件设计上的哲学原理就决定了其非常容易执行矩阵乘法的操作：

- 矩阵乘法存在大量可并行的运算分支，对应成千上万的 threads
- 矩阵乘法的 threads 都遵循相同的运算公式，对应一个 Warp 可以对 32 个 threads 执行相同的指令。
