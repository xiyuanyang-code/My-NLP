## GPU in Depth

### CPUs and GPUs

- CPU 需要强大的计算能力来执行程序生成的二进制代码和汇编指令，需要强大的控制器，并且需要较好的分支预测能力
- GPU 相比于 CPU 具有很多并行的 ALU 算数运行单元，控制器在 CPU 负责协调调度 GPU 的多个 ALU 计算单元。
    - 程序的并行和**吞吐量**是 GPU 运行控制的重点之一

### Atanomy of GPUs

- GPU 存在大量的 SM (Streaming MultiProcessors), 每一个都可以独立的调度多个 SPs (Streaming Processor)，这些可以实现超高进程的并发执行。

    - SP (CUDA Core): 代表并行执行的最基本单元，每一个 SP 可以在一个时钟周期内执行一个时间的计算
    - SM: SM 是由多个 SP 加上其他资源组成的高级处理单元

GPU 同样满足局部性原理，在 SM 内部的 L1 & shared memory 运行速度非常快. (The closer the memory to the SM, the faster it is.)



## Execution Models of GPUs

- Threads: all threads execute the same instructions but with 
different inputs (SIMT).
- Blocks: groups of threads. 每一个 block 运行在一个 SM 上，并有自己独享的内存 (就像一个线程!)
- Warp: Threads always execute in a 'warp' of 32 consecutively numbered threads each.
